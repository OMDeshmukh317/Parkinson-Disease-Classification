## Model Building

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
models=[LogisticRegression(),DecisionTreeClassifier(),RandomForestClassifier(),GaussianNB(),KNeighborsClassifier(),SVC(),XGBClassifier(),AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1),n_estimators=50)]
```

```python
for i in models:
    y_pred=cross_val_predict(estimator=i,X=x,y=y,cv=StratifiedKFold(5),n_jobs=-1)
    print(i)
    print('Confusion matrix:')
    print(confusion_matrix(y,y_pred))
    print('_ _ _ '*5)
    print('Classification Report:')
    print(classification_report(y,y_pred))
    print('_ _ _ '*15)
```

```python
adasyn = SMOTE()
x_n, y_n = adasyn.fit_resample(x, y)
```

```python
for i in models:
    y_pred=cross_val_predict(estimator=i,X=x_n,y=y_n,cv=StratifiedKFold(5),n_jobs=-1)
    print(i)
    print('Confusion matrix:')
    print(confusion_matrix(y_n,y_pred))
    print('_ _ _ '*5)
    print('Classification Report:')
    print(classification_report(y_n,y_pred))
    print('_ _ _ '*15)
```
